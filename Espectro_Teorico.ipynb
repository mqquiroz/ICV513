```python
import numpy as np
import matplotlib.pyplot as plt

g = 9.81 #(m/s2)
A = 5*1e-6
```

```python
df = 0.001 #(Hz)
fmax = 0.5 #(Hz)
f = np.arange(0,fmax+df,df)

E = np.zeros(len(f))

for i in range(len(E)):
  if f[i] >= 0.1:
    E[i] = A*(g**2)*(f[i]**-5)

plt.plot(f,E,'k')
plt.xlim(0,fmax)
plt.ylim(0,50)
plt.grid(linestyle='--',linewidth=0.5)
plt.xlabel('$f$ [Hz]')
plt.ylabel('$E(f)$ [m$^{2}$/Hz]')
plt.savefig('Espectro2.png', dpi=300)
plt.show()
```
