### Representación de figuras en 2D

```python
# Librerias
import numpy as np
import matplotlib.pyplot as plt
```

Se trabajará con una función Gaussiana definida como:

$$f(x,y) = A\cdot e^{-\dfrac{(x-x_{0})^{2} + (y-y_{0})^{2}}{L}}$$

Con $x$ e $y$ en $[0,100]$, $A= 100$, $x_{0} = y_{0} = 50$

```python
#Hacer una figura
alto  = 5
ancho = 5
fig, ax = plt.subplots(figsize=(ancho,alto))

#Datos
A = 100
L = 750
x0 =50
y0 =50

#discretización del dominio
dx = 0.1
dy = dx
x = np.arange(0,100+dx,dx)
y = np.arange(0,100+dy,dy)

X,Y = np.meshgrid(x,y)

#funcion f(x,y)
f = A*np.exp(-( ((X-x0)**2 + (Y-y0)**2)/L))
```

Representación visual de Meshgrid

```python
plt.plot(X[::25,::25],Y[::25,::25],'.k')
plt.show()
```

Gráfico en el plano 2D, los colores representan la función $f(x,y)$

```python
plt.pcolormesh(X,Y,f)
cbar = plt.colorbar()
cbar.set_label('$f(x,y)$')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Representación en 2D a través de curvas de nivel

```python
CS = plt.contour(X,Y,f,levels= np.arange(0,100,20),colors='k')
plt.clabel(CS)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
